<h5 style="color:#0fff9f">3.1 Identificar los valores faltantes</h5>

```python
dfbienes_raices.replace("?", np.nan, inplace = True)
dfbienes_raices.head(5)
```

<h5 style="color:#0fff9f">3.2 Contar la cantidad de valores faltantes</h5>

```python
dfTemp = dfbienes_raices.isnull()
dfTemp.head(5)
```

```python
for column in dfTemp.columns.values.tolist():
    print(column)
    print(dfTemp[column].value_counts())
    print("")
```

```python
valores_faltantes = dfbienes_raices['Serial Number'].isna()
datos_faltantes = dfbienes_raices[valores_faltantes]
print(datos_faltantes)
```

<h5 style="color:#0fff9f">3.3 Imputación de valores numéricos</h5>

```python
av_list_year = dfbienes_raices["List Year"].astype("float").mean(axis=0)
av_list_year = round(av_list_year, 0)
print("La media normalized es : ", av_list_year)
```

```python
dfbienes_raices["List Year"].replace(np.nan, av_list_year, inplace = True)
```

```python
for column in dfTemp.columns.values.tolist():
    print(column)
    print(dfTemp[column].value_counts())
    print("")
```

<h5 style="color:#0fff9f">3.4 Imputación de valores categóricos</h5>

<h5 style="color:#0fff9f">3.5 Eliminar filas con valores faltantes</h5>

```python
dfbienes_raices.dropna(subset =["Serial Number"], axis=0, inplace=True)  
dfbienes_raices.reset_index(drop=True, inplace=True)
dfbienes_raices.dropna(subset =['Date Recorded'], axis=0, inplace=True)  
dfbienes_raices.reset_index(drop=True, inplace=True)
dfbienes_raices.dropna(subset =['Sale Amount'], axis=0, inplace=True)  
dfbienes_raices.reset_index(drop=True, inplace=True)
dfbienes_raices.dropna(subset =['Sales Ratio'], axis=0, inplace=True)
dfbienes_raices.reset_index(drop=True, inplace=True)
dfbienes_raices.dropna(subset =['Address'], axis=0, inplace=True) 
dfbienes_raices.reset_index(drop=True, inplace=True)
```

```python
dfTemp = dfbienes_raices.isnull()
for column in dfTemp.columns.values.tolist():
    print(column)
    print(dfTemp[column].value_counts())
    print("")
```
