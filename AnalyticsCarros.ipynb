<h3 style="color:#5b5990">4. Manejo de valores faltantes</h3>
<p>¿Qué se hace para el manejo de los valores faltantes? </p>
<ol>
    <li>Identificar cómo se muestran los valores faltantes</li>
    <li>Contar la cantidad de valores faltantes en el set de datos</li>
    <li>Imputación de valores numéricos</li>
    <li>Imputación de valores categóricos</li>
    <li>Eliminar filas</li>
</ol>

<h5 style="color:#5b5990">4.1 Identificar los valores faltantes</h5>

```python
dfAutos.replace("?", np.nan, inplace = True)
dfAutos.head(5)
```

<h5 style="color:#5b5990">4.2 Contar la cantidad de valores faltantes</h5>

```python
dfTemp = dfAutos.isnull()
dfTemp.head(5)
```

```python
for column in dfTemp.columns.values.tolist():
    print(column)
    print(dfTemp[column].value_counts())
    print("")
```

<h5 style="color:#5b5990">4.3 Imputación de valores numéricos</h5>

```python
av_normalized = dfAutos["normalized"].astype("float").mean(axis=0)
print("La media normalized es : ", av_normalized)
av_bore = dfAutos["bore"].astype("float").mean(axis=0)
print("La media bore es: ", av_bore)
av_stroke = dfAutos["stroke"].astype("float").mean(axis=0)
print("La media stroke es: ", av_stroke)
av_horsepower = dfAutos["horsepower"].astype("float").mean(axis=0)
print("La media horsepower es: ", av_horsepower)
av_peak = dfAutos["peak"].astype("float").mean(axis=0)
print("La media peak es: ", av_peak)
```

```python
dfAutos["normalized"].replace(np.nan, av_normalized, inplace = True)
dfAutos["bore"].replace(np.nan, av_bore, inplace = True)
dfAutos["stroke"].replace(np.nan, av_stroke, inplace = True)
dfAutos["horsepower"].replace(np.nan, av_horsepower, inplace = True)
dfAutos["peak"].replace(np.nan, av_peak, inplace = True)
dfAutos.head(5)
```

<h5 style="color:#5b5990">4.3 Imputación de valores categóricos</h5>

```python
dfAutos["doors"].value_counts().idxmax()
```

```python
dfAutos["doors"].replace(np.nan, "four", inplace = True)
dfAutos.head(5)
```

<h5 style="color:#5b5990">4.5 Eliminar filas con valores faltantes</h5>

```python
dfAutos.dropna(subset = ["price"], axis=0, inplace=True)
dfAutos.reset_index(drop=True, inplace=True)
```

<h3 style="color:#5b5990">5. Estandarización de Datos</h3>

<ol>
    <li>Verificar el dominio de las variables</li>
    <li>Convertir valores numéricos en función de la interpretación del medio</li>
    <li>Normalizar valores entre 0 y 1</li>
    <li>Creación de rangos de valores (Binning)</li>
    <li>Eliminar filas</li>
</ol>

<h5 style="color:#5b5990">5.1 Verificar el dominio de las variables</h5>

INT ["normalized"], ["horsepower"], ["peak"]   - FLOAT ["bore"], ["stroke"]

```python
dfAutos[["normalized"]] = dfAutos[["normalized"]].astype("int")
dfAutos.dtypes
```

```python
dfAutos[[ "horsepower" , "peak" , "price"]] = dfAutos[[ "horsepower" , "peak" , "price"]].astype("int")
dfAutos.dtypes
```

```python
dfAutos[["bore" , "stroke"]] = dfAutos[["bore" , "stroke"]].astype("float")
dfAutos.dtypes
```

```python
dfAutos.describe()
```

<h5 style="color:#5b5990">5.2 Convertir valores numéricos en función del medio</h5>

```python
dfAutos["City L/100km (Calc)"] = 235/dfAutos["City"]
dfAutos["Highway L/100km (Calc)"] = 235/dfAutos["Higway"]
dfAutos.head(5)
```

<h5 style="color:#5b5990">5.3 Normalizar valores entre 0 y 1</h5>

```python
dfTemp = dfAutos[["City L/100km (Calc)","Highway L/100km (Calc)","engineSize"]]
dfTemp.head(5)
```

```python
dfTemp.describe()
```

```python
dfAutos['City L/100km (norm)'] = dfAutos['City L/100km (Calc)']/dfAutos["City L/100km (Calc)"].max()
dfAutos["Highway L/100km (norm)"] = dfAutos["Highway L/100km (Calc)"]/dfAutos["Highway L/100km (Calc)"].max()
dfAutos["engineSize (norm)"] = dfAutos["engineSize"]/dfAutos["engineSize"].max()

dfTemp = dfAutos[["City L/100km (Calc)","Highway L/100km (Calc)","engineSize","City L/100km (norm)","Highway L/100km (norm)","engineSize (norm)"]]
dfTemp.head(5)
```
